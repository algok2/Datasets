```python
#https://www.kaggle.com/rakannimer/air-passengers
df=pd.read_csv('https://github.com/algok2/Datasets/blob/d275c19fa86abfc292a486851d8f7aa6cb01e8c2/AirPassengers.csv?raw=true')
 
df.head()
```

```python
df.set_index('Month',inplace=True)
df.index=pd.to_datetime(df.index)
 
#drop null values
df.dropna(inplace=True)
 
df.plot()
```

```python
result=seasonal_decompose(df['#Passengers'],  model='multiplicable')
```

```python
result.plot()
```

```python
result.resid.plot()
```
